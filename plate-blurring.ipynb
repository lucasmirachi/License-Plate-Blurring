# Russian Licence Plate Detection & Blurring

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
car_plate1 = cv2.imread("./images/car_plate.jpg")
car_plate2 = cv2.imread("./images/car_plate2.jpeg")
car_plate3 = cv2.imread("./images/car_plate3.jpg")
car_plate4 = cv2.imread("./images/car_plate4.jpg")
```

```python
def display(img):
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    plt.imshow(img)
    plt.show()
```

```python
car_plates = [car_plate1, car_plate2, car_plate3, car_plate4]

for i in range(len(car_plates)):
    display(car_plates[i])
```

```python
plate_cascade = cv2.CascadeClassifier("./haarcascades/haarcascade_licence_plate_rus_16stages.xml")
```

```python
def detect_plate(img):
    
    plate_img = img.copy()
  
    plate_rects = plate_cascade.detectMultiScale(plate_img) 
    
    for (x,y,w,h) in plate_rects: 
        cv2.rectangle(plate_img, (x,y), (x+w,y+h), (0,0,255), 2) 
        
    return plate_img
```

```python
for i in range(len(car_plates)):
    result = detect_plate(car_plates[i])
    display(result)
```

```python
def detect_blur_plate(img):
    
    plate_img = img.copy()
  
    plate_rects = plate_cascade.detectMultiScale(plate_img) 
    
    for (x,y,w,h) in plate_rects: 
        blur_img = cv2.medianBlur((plate_img[y:y+h, x:x+w]), 7)
        plate_img[y:y+h, x:x+w] = blur_img
        
    return plate_img
```

```python
for i in range(len(car_plates)):
    result = detect_blur_plate(car_plates[i])
    display(result)
```
